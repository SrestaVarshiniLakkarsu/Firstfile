### Titanic dataset Visualization 
#### Conclusions

Out of the total passengers, 549 survived while 342 died.

Women were more likely to survive compared to men.

First-class passengers had a higher chance of survival, indicating that rich individuals had better survival prospects.

Most passengers paid lower fares, especially those in third class.
There's no clear correlation between fare paid and survival rate within each passenger class.

Younger passengers, particularly children, had a higher chance of survival, whereas elderly passengers had lower survival rates.

Survival rates varied based on the embarkation port, likely reflecting differences in socio-economic backgrounds of passengers boarding from different ports.

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='class', y='survived', hue='sex')
plt.title('Survival Rate by Passenger Class and Gender')
plt.show()
```

Passenger Class: First-class passengers had a higher survival rate, indicating socio-economic status played a role in survival chances.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', kde=True, hue='sex')
plt.title('Age Distribution by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='embarked', hue='survived')
plt.title('Survival Count based on Embarkation Port')
plt.show()
```

Embarkation Port: The survival count varied based on the embarkation port, potentially reflecting the socio-economic distribution of passengers from these ports. C port passengers have grater survival rate.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='class', y='fare', hue='survived')
plt.ylim(0, 300)  # Limiting y-axis to 300 for better visualization
plt.title('Fare distribution by Passenger Class and Survival')
plt.show()
```

Fare Distribution: The majority of passengers paid lower fares, are  with the larger number of third-class tickets.
Fare and Survival: Within each passenger class, there wasn't a consistent pattern to suggest that higher fares directly led to better survival chances.
